```python
musics["top genre"] = musics["top genre"].apply(lambda x: "Hip hop" if "hip hop" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Hip hop" if "rap" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Hip hop" if "hollywood" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Rock" if "rock" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Rock" if "permanent wave" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Rock" if "indie" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Electronic" if "electronic" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Electronic" if "electro" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Electronic" if "house" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Electronic" if "edm" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Electronic" if "belgian edm" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Electronic" if "complextro" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Electronic" if "downtempo" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Electronic" if "brostep" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Electronic" if "escape room" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Electronic" if "big room" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Pop" if "pop" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Pop" if "australian dance" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Pop" if "boy band" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Pop" if "neo mellow" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Pop" if "irish" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Soul" if "soul" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Soul" if "country" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Soul" if "pop" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Soul" if "r&b" in x else x)
musics["top genre"] = musics["top genre"].apply(lambda x: "Latin" if "latin" in x else x)
```

## Existe uma correlação entre o quanto uma música é fácil de se dançar e o quanto ela é alegre/positiva ? Se sim, como ela se comporta?

Com base nas análises no notebook par desse, é concluído que existe sim uma correlação.

```python
corr_violence_romance = musics["val"].corr(musics["dnce"])
corr_violence_romance.round(decimals = 4)
```

Como visto acima, a postividade de uma música tem uma correlação com a dançabilidade dessas mesmas, no caso (0.5017).

Não somente, é possível visualizar isso no gráfico abaixo:

```python
sns.scatterplot(data = musics, x = musics["val"], y = musics["dnce"], alpha = 0.8, color = "darkblue")
```

    Observando o gráfico, pode-se perceber a relação de uma característica com a outra, todavia, pelo índice de correlação (0.5017) não ser tão significativo, é concluído que é mais fácil e provável de se dançar uma música que trata sobre conteúdo positivo/alegre, porém, não é o único fator que influencia nesse aspecto.

O qual é definido majoritariamente por dois aspectos, o primeiro trata sobre as escalas musicais utilizadas, e o outro tem relação direta sobre as palavras cantadas na música em questão.

Para tal parâmetro, foi utilizado os índices de alegria e positividade do dataset, que seguem a seguinte escala:

+ 80 <= índice <= 100   = Extremamente alegre
