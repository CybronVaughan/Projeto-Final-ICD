    Índices de correlação entre as características musicais

Aqui é possível analisar quais características musicais tem mais/menos correlação entre si

```python
corr = df.corr().round(decimals = 4)
corr
```

```python
#corr.style.background_gradient(cmap='coolwarm').set_properties(**{'font-size': '0pt'})
corr.style.background_gradient(cmap='coolwarm')
```

Aqui foi feita a análise dos índices com a exclusão da diagonal principal, a qual faz correlação consigo mesma, logo não faria esntido analisar esses dados. No caso, ela foi preenchida com os valores de 0, após isso, foi feito a análise a partir dos novos valores.

```python
np.fill_diagonal(corr.values, 0)
corr
```

Índice de correlação entre a acústica e a energia das músicas

```python
corr_violence_romance = df["acousticness"].corr(df["energy"])
corr_violence_romance.round(decimals = 4)
```

Acima é visto um exemplo de como duas características musicais se correlacionam, a acústica de uma música tem uma forte correlação negativa com a energia da música, no caso igual a ( -0.72 ), ou seja, quanto mais uma música é considerada energética, menos tende ser a sua acústica!
