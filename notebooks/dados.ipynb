```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

    LENDO OS ARQUIVOS CSV E OS TRANSFORMANDO EM DATAFRAMES

```python
data = pd.read_csv(r"/home/vinicius_olzon/Documents/Projeto-Final-ICD/tcc_ceds_music.csv")
df = pd.DataFrame(data)
```

```python
df.head(3)
```

    LIMPEZA DOS  DADOS


```python
del df["Unnamed: 0"]
del df["len"]
del df["age"]
df.isnull().values.any()
```

    MEDIDAS DE CENTRALIDADE E DISPERSÃO

```python
df.describe()
```

    LINEPLOT ENTRE ANO DE LANÇAMENTO CORRELACIONADO COM A OBSCENDIDADE DAS MÚSICAS

```python
sns.set_theme()
sns.lineplot(x="release_date", y="obscene", data= df)
```

    BARPLOT ENTRE OS GÊNEROS MUSICAIS E A OBSCENIDADE DAS MÚSICAS

```python
sns.barplot(x="genre", y="obscene", data= df)
```

DISPLOT DOS GÊNEROS

```python
sns.displot(x="genre", data=df, hue = "genre",alpha = 0.9)
```

    BARPLOT DOS 10 ARTISTAS COM MAIS MÚSICAS 

```python
most_artist = df.value_counts("artist_name")
most5_artist = most_artist.head(10)
most5df = pd.DataFrame(most5_artist)
most5df.index
most5_artist.unique()

sns.barplot(x = most5df.index, y = most5_artist.unique(), data =df, alpha = 1)
plt.xticks(rotation = 70)
```

    BOXPLOT SOBRE DE CADA MÚSICA EM SEGUNDOS

```python
sns.boxplot(y = "violence",data = df)
```

    BOXPLOT ENTRE GÊNEROS MUSICAIS E A VIOLÊNCIA

```python
sns.boxplot(data = df, x = 'genre', y = 'violence', whis = 2, width  = 0.7)
```

    ÍNDICES DE CORRELAÇÃO ENTRE AS CARACTERÍSTICAS MUSICAIS

```python
corr = df.corr().round(decimals = 4)
corr
```

    ÍNDICE DE CORRELAÇÃO ENTRE A ACUSTICIDADE E A ENERGIA DAS MÚSICAS

```python
corr_violence_romance = df["acousticness"].corr(df["energy"])
corr_violence_romance.round(decimals = 4)
```

```python
corr.style.background_gradient(cmap='coolwarm')
```

```python
corr.style.background_gradient(cmap='coolwarm').set_properties(**{'font-size': '0pt'})
```

        Análise dos índices de correlação com a exclusão da diagonal principal, a qual faz correlação consigo mesma. No caso, ela foi preenchida com os valores de 0, após isso, foi feito a análise a partir dos novos valores.

```python
np.fill_diagonal(corr.values, 0)
corr
```
